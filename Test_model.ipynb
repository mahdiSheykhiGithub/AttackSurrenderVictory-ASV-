# Imports

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

# Test

```python
model = YOLO('./model/best.pt')
```

```python
model.names
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ans, frame =cap.read()

    if ans:
        results = model.predict(frame, conf=0.841, show=False)
        results_frame = results[0].plot()
        cv2.imshow('camera', results_frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
